```python
#create a new column in the df (Gender) and set equal to either M or F
df['Gender'] = df['Sex'].map( lambda x: x[0].upper() )

#let's instead make it 0 for female, 1 for male
df['Gender'] = df['Sex'].map( {'female':0,'male':1}).astype(int)

df.head()
```

```python
#calculate median ages of passenger class and gender
median_ages = np.zeros((2,3))

for i in range(0,2):
    for j in range(0,3):
        median_ages[i,j] = df[(df['Gender']==i) & (df['Pclass']==j+1)]['Age'].dropna().median()

median_ages
```

```python
#now start process of filling NaN values, but we don't want simply to fill and forget
df['AgeFill'] = df['Age']

df[df['Age'].isnull()][['Gender','Pclass','Age','AgeFill']].head(10)
```

```python
#fill in those with a missing Age value by taking the median by class and gender for those
for i in range(0,2):
    for j in range(0,3):
        df.loc[ (df.Age.isnull()) & (df.Gender == i) & (df.Pclass == j+1), 'AgeFill'] = median_ages[i,j]
        
df.head(10)
```

```python
#create a feature that records whether Age was null and had to be filled via median
df['AgeIsNull'] = pd.isnull(df.Age).astype(int)
df[df.Age.isnull()].head()
```

```python
#feature engineering: FamilySize as a combination of SibSp (siblings and spouses) and Parch (parents and children)
df['FamilySize'] = df['SibSp'] + df['Parch']

#potential feature: multiply age by class
df['Age*Class'] = df.Age * df.Pclass
```
